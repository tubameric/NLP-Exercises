```python
# Code snippet 1
# Importing the reviews dataset
url='https://raw.githubusercontent.com/nvamsimohan/DallasDSA/main/Dallas%20house%20yelp%20reviews%20(2).txt'

url
```

```python
# Code snippet 2
import requests
reviews = requests.get(url)

text = reviews.text

text
```

```python
# Code snippet 3
# Importing the relevant NLP packages
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
# Code snippet 4
# Initializing vader sentiment analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Code snippet 5
# Calling polarity scores to do the sentiment analysis
scores = sia.polarity_scores(text)

scores
```
